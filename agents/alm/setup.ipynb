The model takes the current price state and decides whether to perform rebalancing or not. It only considers the wide range and does not take into account the volatile token/stable token ratio, focusing solely on yield. It assumes that all liquidity is allocated to the wide range.

The model expects the following input (here, price refers to the actual volatile token price in terms of the stable token, not in ticks):

np.array([
    self.price_normed,
    self.price_mean_normed,
    self.price_vol_normed,
    self.price_slope_normed
])

Here, the price (and mean price) is normalized as:

$$
\text{price normed} = \frac{\text{price} - \text{position price lower}}{2*\text{wide range wide}}\,.
$$

Volatility is normalized as:

$$
\text{volatility normed} = \text{volatility}*0.00025\,.
$$

The slope is calculated using the least squares method and normalized as:

$$
\text{slope normed} = \frac{1}{2}\left(\frac{2}{\pi}\arctan{(0.5*\text{slope})} + 1\right)\,.
$$

Both volatility and slope are computed over the time horizon specified in the configs.

Model outputs are follows:

0 - do nothing\;

1 - do rebalance\;
